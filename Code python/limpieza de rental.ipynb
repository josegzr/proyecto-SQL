# 1. limpieza de datos del csv rental

```python
import pandas as pd 
import numpy as np
```

```python
data = pd.read_csv('../Data/rental.csv', encoding='latin1')
data
```

```python
data['rental_id'] = range(1,len(data)+1)
```

```python
data
```

```python
data[['Fecha', 'Hora_rental']] = data['rental_date'].str.split(' ', expand=True)
data
data[['year_rental', 'Mes_rental', 'Dia_rental']] = data['Fecha'].str.split('-', expand=True)
data
```

```python
data.year_rental.unique()
```

```python
data_null=[ 'year_rental','rental_date','Fecha']
data.drop(data_null, axis=1,inplace=True)
```

```python
data
```

```python
data[['Fecha', 'Hora_return']] = data['return_date'].str.split(' ', expand=True)
data
data[['year_return', 'Mes_return', 'Dia_return']] = data['Fecha'].str.split('-', expand=True)
data
```

```python
data_null=[ 'year_return','return_date','Fecha','last_update']
data.drop(data_null, axis=1,inplace=True)
```

```python
data
```

```python
data= data[['rental_id','inventory_id',  'customer_id', 'staff_id', 'Mes_rental', 'Dia_rental', 'Hora_rental','Mes_return','Dia_return','Hora_return']]
data
```
