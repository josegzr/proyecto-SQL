# 1. limpieza de datos del csv inventory

```python
import pandas as pd 
import numpy as np
```

```python
data = pd.read_csv('../Data/inventory.csv', encoding='latin1')
data
```

```python
data.isna().sum()
```

```python
data[['Fecha', 'Hora']] = data['last_update'].str.split(' ', expand=True)
data
data[['year_update', 'Mes_update', 'Dia_update']] = data['Fecha'].str.split('-', expand=True)
data
```

```python
data_null=['Fecha', 'last_update']
data.drop(data_null, axis=1,inplace=True)
```

```python
data= data[['inventory_id','film_id', 'store_id', 'year_update', 'Mes_update', 'Dia_update', 'Hora' ]]
```

```python
data
```

```python
data=data.drop_duplicates()
```

```python
data.columns = ['inventory_id','film_id', 'store_id', 'year_update', 'Mes_update', 'Dia_update', 'Hora' ]
```
