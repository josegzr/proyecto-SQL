# 1. limpieza de datos del csv language

```python
import pandas as pd 
import numpy as np
```

```python
data = pd.read_csv('../Data/language.csv', encoding='latin1')
data
```

```python
data[['Fecha', 'Hora']] = data['last_update'].str.split(' ', expand=True)
data
data[['year', 'Mes', 'Dia']] = data['Fecha'].str.split('-', expand=True)
data
```

```python
data_null=['Fecha', 'last_update']
data.drop(data_null, axis=1,inplace=True)
```

```python
data= data[['language_id','name', 'year', 'Mes', 'Dia', 'Hora' ]]
```
