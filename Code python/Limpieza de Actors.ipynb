# 1. limpieza de datos del csv actors

```python
import pandas as pd 
import numpy as np
```

```python
data = pd.read_csv('../Data/actor.csv', encoding='latin1')
data
```

```python
data.info()
```

```python
data.isna().sum()
```

```python
duplicados=data.duplicated(subset=['first_name', 'last_name'])
data[duplicados].index[0]
```

```python
data.drop(109, inplace=True)
```

```python
data.reset_index(drop=True, inplace=True)
```

```python
data['actor_id'] = range(1,len(data)+1)
```

```python
data.shape
```

```python
data.last_update.head(100)
```

```python
data=data.drop_duplicates()
```

```python
data
```

```python
data[['Fecha', 'Hora']] = data['last_update'].str.split(' ', expand=True)
data
data[['year_update', 'Mes_update', 'Dia_update']] = data['Fecha'].str.split('-', expand=True)
data
```

```python
data_null=['Fecha', 'last_update']
data.drop(data_null, axis=1,inplace=True)
```

```python
data
```

```python
data= data[['actor_id','first_name', 'last_name', 'year_update', 'Mes_update', 'Dia_update', 'Hora' ]]
```

```python
data['complete_name'] = data['first_name'] + ' ' + data['last_name']
```

```python
data_null=['first_name', 'last_name']
data.drop(data_null, axis=1,inplace=True)
```

```python
data= data[['actor_id','complete_name', 'year_update', 'Mes_update', 'Dia_update', 'Hora' ]]
```

```python
data
```

```python
data_null=[ 'Hora']
data.drop(data_null, axis=1,inplace=True)
```
